```python
import collections

import tensorflow as tf
import tensorflow_probability as tfp
tfd = tfp.distributions

try:
  tf.compat.v1.enable_eager_execution()
except ValueError:
  pass

import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# A standard normal
normal = tfd.Normal(loc=2., scale=1.)
print(normal)
```

```python
# Plot 1000 samples from a standard normal
samples = normal.sample(1000)
sns.distplot(samples)
plt.title("Samples from a standard Normal")
plt.show()
```

```python
# Bernauli Distribution

b_dist_arr = tfd.Bernoulli(probs=[.3, .5, .7])

b_sample = b_dist_arr[0].sample(100)
#a1_sample = a_dist_arr[1].sample(100)
print('b_sample 1', b_sample)
print('b_sample 1', b_dist_arr[1].sample(100))
print('b_sample 1', b_dist_arr[2].sample(100))


print('b_dist_arr ', b_dist_arr)
```

```python
# Poission Distribution

p_dist_arr = tfd.Poisson(rate=[1., 4., 10.])

p_sample = p_dist_arr[0].sample(100)
p_sample1 = p_dist_arr[1].sample(100)
p_sample2 = p_dist_arr[2].sample(100)


#a1_sample = a_dist_arr[1].sample(100)
print('p_sample 1', p_sample)
print('p_sample 1', p_dist_arr[1].sample(100))
print('p_sample 1', p_dist_arr[2].sample(100))

sns.distplot(p_sample)
sns.distplot(p_sample1)
sns.distplot(p_sample2)

#sns.distplot(a1_sample)

plt.title("Samples from a standard Normal")
plt.show()


print('p_dist_arr ', p_dist_arr)
```

```python

dist_normal = tfd.Normal(loc=[-1., 1], scale=[0.1, 0.5])



# Make independent distribution from a 2-batch Normal.
ind_orig = tfd.Independent(
    distribution= dist_normal)

print('ind_orig batch_shape', ind_orig.batch_shape, ' event_shape ', ind_orig.event_shape)

ind_new_dims = tfd.Independent(
    distribution= dist_normal,
    reinterpreted_batch_ndims=1)

# All batch dims have been 'absorbed' into event dims.

print('ind_new_dims batch_shape', ind_new_dims.batch_shape, ' event_shape ', ind_new_dims.event_shape)


# Make independent distribution from a 2-batch bivariate Normal.
ind = tfd.Independent(
    distribution=tfd.MultivariateNormalDiag(
        loc=[[-1., 1], [1, -1]],
        scale_identity_multiplier=[1., 0.5]),
    reinterpreted_batch_ndims=1)

# All batch dims have been 'absorbed' into event dims.
ind.batch_shape  # ==> []
ind.event_shape  # ==> [2, 2]
```
